```python
import torch
import torchvision
from torchvision.transforms import ToTensor
import matplotlib.pyplot as plt
import numpy as np
```

```python
train_ds=torchvision.datasets.MNIST('data/',train=True,transform=ToTensor(),download=True)
test_ds=torchvision.datasets.MNIST('data/',train=False,transform=ToTensor(),download=True)
```

```python
train_dl=torch.utils.data.DataLoader(train_ds,batch_size=64,shuffle=True,num_workers=4)#num_workers为子线程的数量
test_dl=torch.utils.data.DataLoader(test_ds,batch_size=64,num_workers=4)
```

```python
imgs,labels=next(iter(train_dl))
print(imgs.shape)
print(labels.shape)
```

```python
plt.figure(figsize=(10,1))
for i,img in enumerate(imgs[:10]):
    npimg=img.numpy()
    npimg=np.squeeze(npimg)
    plt.subplot(1,10,i+1)
    plt.imshow(npimg)
    plt.axis('off')
```
